```python
def print_name(name , n):

  if n == 0:
    return 

  print(name + ' ' + str(n) + ',')
  print_name(name , n-1)
```

```python
print_name('pranjal' , 5)
```

```python
def printVal_1_N(i,n):
  print(i)

  if(i>=n):  ## stop at n 
    return 

  printVal_1_N(i+1,n)
```

```python
printVal_1_N(1,5)
```

```python
def printVal_N_1(i,n):

  print(i)

  if(i<=n):
    return 

  printVal_N_1(i-1,n)
```

```python
printVal_N_1(5,1)
```

#### backtracking 

```python
def printVal_1_N(i,n):
  ## i+ 1 not allowed , print while returning 

  if i < n:
    return 

  printVal_1_N(i-1,n)
  print(i)
```

```python
printVal_1_N(5,1)
```

```python
def PrintVal_N_1(i,n): ## i -1 not allowed: Input (1,5)
  
  if i > n:
    return

  PrintVal_N_1(i+1,n)
  print(i)
```

```python
PrintVal_N_1(3,10)
```
